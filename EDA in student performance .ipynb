  ####   We can see Value counts for each Male and Female in the dataset

```python
df['gender'].value_counts().plot.pie(figsize=(9,9),autopct='%.2f%%',shadow=True,explode=[0.0,0.1],fontsize=13,label=('female','male'),title='gender',colors=['pink','royalblue']);
```

 
 #### we can also plot dataset with pandas library like that easily and quickly

 ####  Value counts for lunch column in the dataset with pandas library

```python
df['lunch'].value_counts().plot.bar(color=['orange','black'])
plt.xticks(rotation=45)
plt.show()
```

 #### Value counts for test preparation course column in the dataset with pandas library

```python
df['test preparation course'].value_counts().plot.bar(color=['red','green'])
plt.xticks(rotation=45)
plt.show()
```

#### You can make a new column and add it to the dataset
#### Total score that is the new column consest of the sum to the score

```python
df['total score'] = df['math score'] +df['reading score'] +df['writing score']
```

```python
df.head()
```

#### Another way to see the score VS gender 

```python
df.groupby('gender').mean()[['math score','reading score','writing score']].transpose().plot(kind='bar',figsize=(9,6))
```

 #### Correlation

 #####  You can notice the correlation between columns

```python
c= df.corr()
sns.heatmap(c,annot=True)
c
```

 ####  There is  high correlation between reading,writing and total score

```python
sns.heatmap(pd.pivot_table(data=df , index='gender', columns='lunch',values='math score').T,annot=True)
```

 #### You can notice that male and standard student have high score in math
 #### But female and free student have low score in math

```python
sns.heatmap(pd.pivot_table(data=df , index='gender', columns='lunch',values='writing score').T,annot=True,fmt='.2f')
```

#### You can notice that female and standard student have high score in writing
#### But male and free student have low score in writing
